```python
import cv2
def frame_faces_with_emotion_label_from_video(
    video_path,
    faces_and_rectangles_from_image, 
    emotion_and_likeliness_from_face, 
):
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)
    webcam_shape = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH)) , int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
    if not video_capture.isOpened():
        print(f"Cannot open camera {video_capture = }")
        return
    while True:
        ok, webcam_frame = video_capture.read()
        if not ok:
            print(f"end of video stream {ok = }")
            break
        faces, rectangles = None, None
        emotion_likeliness_couples = None
        try:
            faces, rectangles = faces_and_rectangles_from_image(webcam_frame)
        except Exception as error:
            print(error)
        if faces:
            emotion_likeliness_couples = list(
                emotion_and_likeliness_from_face(face)
                for face in faces
            )
        if rectangles:
            list(
                cv2.rectangle(webcam_frame, top_left, bottom_right, color=(255, 0, 0), thickness=2) 
                for top_left, bottom_right in rectangles
            )            
        if emotion_likeliness_couples and rectangles:
            for (emotion, likeliness), rectangle in zip(emotion_likeliness_couples, rectangles):
                left, top  = rectangle[0]
                cv2.putText(
                    img=webcam_frame,
                    text=f"{emotion} {likeliness}",
                    org=(left+3,top+20),
                    # org=(30,30),# bottomLeftCornerOfText,
                    fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale=1,
                    color=(255,0,255),
                    thickness=3,
                    bottomLeftOrigin=False,
                )
        cv2.imshow('webcam_frame', webcam_frame) # showing webcam output
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    video_capture.release()
    cv2.destroyAllWindows()
    
frame_faces_with_emotion_label_from_video(
    video_path=video.find_webcam_index(),
    faces_and_rectangles_from_image=faces_and_rectangles_from_image,
    emotion_and_likeliness_from_face=emotion_and_likeliness_from_face,
)
```
